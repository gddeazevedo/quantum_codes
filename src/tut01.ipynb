```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator, Statevector
```

```python
n_qubits = 1
n_cbits  = 1

qubit = 0 # the index of the qubit in the circuit
cbit  = 0 # the index of the classical bit

qc = QuantumCircuit(n_qubits, n_cbits)

qc.reset(0)
qc.x(0)
qc.measure(qubit, cbit) # measure qubit index 0 and store the result in the classical bit index 0

qc.draw()
```

```python
qc_x = QuantumCircuit(1)
qc_x.x(0)
X = Operator(qc_x)
X.draw('latex')
```

```python
X = Operator.from_label('X')
ket_0 = Statevector.from_label('0')
ket_0.evolve(X).draw('latex')
```

```python
ket_1 = Statevector.from_label('1')
ket_1.evolve(X).draw('latex')
```

```python
H = Operator.from_label('H')
ket_0.evolve(H).draw('latex')
```

```python
ket_1.evolve(H).draw('latex')
```

```python
H.draw('latex')
```

```python
from qiskit.providers.basic_provider import BasicSimulator


sim = BasicSimulator()
n_qubits = 1
n_cbits  = 1

qubit = 0 # the index of the qubit in the circuit
cbit  = 0 # the index of the classical bit

qc = QuantumCircuit(n_qubits, n_cbits)

qc.h(0)
qc.measure(qubit, cbit) # measure qubit index 0 and store the result in the classical bit index 0

qc.draw()
```

```python
job = sim.run(qc, shots=100)
result = job.result()
counts = result.get_counts()
print(counts)
```
