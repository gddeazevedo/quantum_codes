```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
```

```python
qubits = [0, 1, 2]
cbits  = [0, 1, 2]
n_quibits = 3
n_cbits = 3

qc = QuantumCircuit(n_quibits, n_cbits)

qc.h(qubits)
qc.measure(qubits, cbits)
qc.draw()
```

```python
Statevector(qc).draw("latex")
```
